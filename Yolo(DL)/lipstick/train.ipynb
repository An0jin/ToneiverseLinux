```python
from ultralytics import YOLO
import koreanize_matplotlib
```

```python
model=YOLO('yolo26n.pt')
model.train(
    data='./dataset/data.yaml',
    epochs=300,
    imgsz=640,
    batch=32,
    val=True,
    # --- 색상 관련: 기존 유지 ---
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    # --- 구도 관련: 대폭 강화 ---
    degrees=15.0,      # 회전각을 조금 더 확보
    translate=0.3,     # 0.1 -> 0.3으로 강화 (객체를 구석으로 강제 이동)
    scale=0.9,         # 0.5 -> 0.9로 강화 (범위: 0.1 ~ 1.9배. 아주 작은 객체 생성 유도)
    perspective=0.001, # 미세한 원근감 추가 (다양한 각도 대응)
    fliplr=0.5,
    flipud=0.2,        # 립스틱은 뒤집혀 있어도 인식해야 하므로 추가
    mosaic=1.0,        # 필수 유지
    mixup=0.1,
    copy_paste=0.1,    # 객체를 복사해서 다른 배경에 붙임 (중앙 집중 해소에 효과적)
    patience=50,
)
```

```python
model=YOLO('yolo26n.pt')
model.train(
    data='./dataset/data.yaml',
    epochs=300,
    imgsz=640,
    batch=32,
    val=True,
    # --- 색상 정밀도 극대화 ---
    hsv_h=0.01,        # 0.0보다는 미세하게 주어 조명 변화 대응 (0.015보다 더 보수적으로 하향)
    hsv_s=0.05,        # 채도 변화도 아주 최소화
    hsv_v=0.15,        # 밝기 변화는 어느 정도 허용
    # --- 기하학적 변형 완전 차단 ---
    degrees=0.0,
    translate=0.0,     # 위치 이동 금지
    scale=0.0,         # 크기 조절 금지 (픽셀 무결성 유지)
    fliplr=0.5,        # 좌우 반전만 허용
    # --- 사용자 지적 반영 ---
    mosaic=0.0,        # 0.0으로 설정하여 얼굴 픽셀 밀도 유지
    mixup=0.0,         # 색 섞임 방지
    patience=50,
)
```
