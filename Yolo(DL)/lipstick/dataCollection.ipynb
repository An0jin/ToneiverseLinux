```python
import os
import shutil
from roboflow import Roboflow
rf=Roboflow(api_key=os.getenv("roboflow"))
rf
```

```python
links=['https://universe.roboflow.com/zaabtahi/lipstick-k4a0p',
 'https://universe.roboflow.com/mateo-ojeda/lipstick-kimjy',
 'https://universe.roboflow.com/nathacha-wamta-mp74y/lipstick-tywhr',
 'https://universe.roboflow.com/passcheckers/lipstick-vanyi',
 'https://universe.roboflow.com/create-space/lipstick-sk2se',
 'https://universe.roboflow.com/projectaie223objectdetection/lipstick-lfcgb',
 'https://universe.roboflow.com/elmodata/lipstick-dataset',
 'https://universe.roboflow.com/bridger-sasha-nuy0j/lipstick-packaging',
 'https://universe.roboflow.com/akash-rajinikanthan-mvsrr/lipstick-identification',
 'https://universe.roboflow.com/adinda-salsabila/lipstick-detection']
```

```python
for link in links:
    linklist=link.split("/")
    version=1
    while True:
        try:
            rf.workspace(linklist[-2]).project(linklist[-1]).version(version).download("folder")
            version+=1
        except:
            try:
                rf.workspace(linklist[-2]).project(linklist[-1]).version(version).download("yolov8")
                version+=1
            except:
                break
```
